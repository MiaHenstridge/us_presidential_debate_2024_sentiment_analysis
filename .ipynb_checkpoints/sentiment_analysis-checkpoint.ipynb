## Load transcript

```python
# read the transcript (line by line)
transcript = []

with open('./debate_transcript.txt', 'r') as file:
    for line in file:
        if line.strip():    # skip the empty lines
            transcript.append(line.rstrip())
```

## Data preprocessing

### Step 1: Extract relevant parts of the data

```python
# separate transcript by each candidate
harris_transcript = [line for line in transcript if line.startswith('VICE PRESIDENT KAMALA HARRIS:')]
trump_transcript = [line for line in transcript if line.startswith('FORMER PRESIDENT DONALD TRUMP:')]
```

```python
# extract only the speeches of each candidate

# harris
start_pattern = r'VICE PRESIDENT KAMALA HARRIS:\s+'
harris_transcript = [re.sub(start_pattern, '', line) for line in harris_transcript]

# trump
start_pattern = r'FORMER PRESIDENT DONALD TRUMP:\s+'
trump_transcript = [re.sub(start_pattern, '', line) for line in trump_transcript]
```

```python
# combine the speeches in the transcript into one large block for each speaker
harris_text = ' '.join(harris_transcript).strip()
trump_text = ' '.join(trump_transcript).strip()
```

```python
harris_text[:200]
```

```python
trump_text[:200]
```

## Step 2: Further necessary text preprocessing

```python
# load stop words
stop_words = set(stopwords.words('english'))

# function to clean text
def clean_text(text):
    # lowercase the text
    text = text.lower()
    # remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # tokenization
    tokens = word_tokenize(text)
    # remove stop words
    filtered_tokens = [word for word in tokens if word not in stop_words]
    # join tokens back to string
    return " ".join(filtered_tokens)
```

```python
# cleaned text
harris_text_cleaned = clean_text(harris_text)
trump_text_cleaned = clean_text(trump_text)
```

```python
harris_text_cleaned[:200]
```

## Step 3: Overall Sentiment and Emotion Analysis

#### Overall subjectivity & polarity

```python
# Analyze subjectivity / polarity 
def get_subjectivity_polarity_scores(text):
    sentiment = TextBlob(text).sentiment
    polarity, subjectivity = sentiment.polarity, sentiment.subjectivity
    return polarity, subjectivity
```

```python
harris_polarity, harris_subjectivity = get_subjectivity_polarity_scores(harris_text_cleaned)
trump_polarity, trump_subjectivity = get_subjectivity_polarity_scores(trump_text_cleaned)

print(f'Harris - polarity score: {harris_polarity:.4f}; subjectivity score: {harris_subjectivity:.4f}')
print(f'Trump - polarity score: {trump_polarity:.4f}; subjectivity score: {trump_subjectivity:.4f}')
```

It seems that Trump's speeches display more subjectivity than Harris'. Let's use a different analyzer to gain more insights into the positivity / negativity of the sentiment.

#### Overall positive / negative / neutral polarity

```python
# create an instance of NLTK's sentiment intensity analyzer
sia = SentimentIntensityAnalyzer()
```

```python
# Analyze sentinment intensity 
def get_sentiment_intensity(text):
    polarity_scores = sia.polarity_scores(text)
    return polarity_scores
```

```python
harris_sia = get_sentiment_intensity(harris_text_cleaned)
trump_sia = get_sentiment_intensity(trump_text_cleaned)

print(f'Harris sentiment intensity scores: {harris_sia}')
print(f'Trump sentiment intensity scores: {trump_sia}')
```

The results show that Harris displays more positivity in her speeches compared to Trump.

#### Emotion scores

```python
# measure emotion scores and normalize them
def get_emotion_scores(text):
    # create an NRCLex object
    nrc = NRCLex(text)
    # affect_dict returns the list of emotion-related words and their associated emotion/sentiment
    word_emo = nrc.affect_dict
    # count occurrences of each word in the text
    word_cnt = Counter(text.split())
    # initialize weighted emotion counts to 0
    weighted_emotion_counts = {emotion: 0 for emotion in nrc.raw_emotion_scores}
    # loop through the words in the text and distribute the counts proportionally
    for wd, emo_ls in word_emo.items():
        if emo_ls:
            # if the word is associated with emotions, get the word count
            occurrences = word_cnt[wd]
            # and distribute the weight across the emotions
            weight = occurrences / len(emo_ls)
            for e in emo_ls:
                weighted_emotion_counts[e] += weight
    # calculate the total weight
    total_weight = sum(weighted_emotion_counts.values())
    # normalize the weighted emotion counts
    if total_weight > 0:
        normalized_scores = {e: cnt/total_weight for e, cnt in weighted_emotion_counts.items()}
    else:
        normalized_scores = {e: 0 for e, cnt in weighted_emotion_counts.items()}
    return normalized_scores
```

```python
harris_emotion_scores = get_emotion_scores(harris_text_cleaned)
trump_emotion_scores = get_emotion_scores(trump_text_cleaned)

emotion_scores = pd.concat([pd.Series(harris_emotion_scores), pd.Series(trump_emotion_scores)], axis=1)
emotion_scores.columns = ['Harris', 'Trump']

emotion_scores.sort_index()
```

```python
# plot noramlized emotion scores for the 2 candidates
ax = emotion_scores.sort_index().plot(kind='bar', color=['blue', 'red'])
plt.title('Emotion Scores (Normalized)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('emotion_scores_normalized.png')
plt.show()
```

The top 3 emotions for both candidates are the same: `trust`, `anticipation`, `fear`.

We also see once again that Harris scores higher than Trump in terms of `positive` sentiment.

### Step 4: Sentiment analysis over time

Here, instead of analyzing sentiment measured over the entire transcript of each candidate, we will measure them speech by speech.

```python
# Since we are not going to merge all the speech into 1 large text block for each candidate,
# we need to replicate the text cleaning steps for each speech
# load stop words
stop_words = set(stopwords.words('english'))

# function to clean speeches
def clean_speech(speech_list):
    # list to store processed speeches
    speech_list_cleaned = []
    for s in speech_list:
        # lowercase the text
        s = s.lower()
        # remove punctuation
        s = s.translate(str.maketrans('', '', string.punctuation))
        # tokenization
        tokens = word_tokenize(s)
        # remove stop words
        filtered_tokens = [word for word in tokens if word not in stop_words]
        # join tokens back to string and append to list of cleaned speech
        speech_list_cleaned.append(" ".join(filtered_tokens))
    
    return speech_list_cleaned
```

```python
# cleaned sentences
harris_speech_cleaned = clean_speech(harris_transcript)
trump_speech_cleaned = clean_speech(trump_transcript)
```

### 4.1. Subjectivity & polarity over time

```python
# function to extract subjectivity and polarity score sentences by sentence
def get_subjectivity_polarity_over_time(speech_list):
    sub = []
    pol = []

    for s in speech_list:
        blob = TextBlob(s).sentiment
        sub.append(blob.subjectivity)
        pol.append(blob.polarity)
    return pol, sub
```

```python
harris_pol, harris_sub = get_subjectivity_polarity_over_time(harris_speech_cleaned)
trump_pol, trump_sub = get_subjectivity_polarity_over_time(trump_speech_cleaned)
```

```python
# plot polarity and subjectivity over time
for harris, trump, sentiment in zip((harris_pol, harris_sub), (trump_pol, trump_sub), ('Polarity', 'Subjectivity')):
    fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 5))

    ax[0].plot(harris, color='b', label='Harris')
    ax[0].set_ylabel('Score')
    ax[0].set_xlabel('Speech')
    ax[0].set_title(f'{sentiment} over time')
    ax[0].legend(loc='upper right')
    
    ax[1].plot(trump, color='red', label='Trump')
    ax[1].set_ylabel('Score')
    ax[1].set_xlabel('Speech')
    ax[1].legend(loc='upper right')

    if sentiment == 'Subjectivity':
        ax[0].set_ylim(-1.1, 1.1)
        ax[1].set_ylim(-1.1, 1.1)
    else:
        ax[0].set_ylim(-0.1, 1.1)
        ax[0].set_ylim(-0.1, 1.1)
        
    plt.tight_layout()
    plt.savefig(f'{sentiment}_over_time.png')
    plt.show()
```

Harris' polarity score remains stable around 0 to 0.5, showing more neutrality while Trump's polarity score display a lot of fluctutations. This shows that Trump is more emotionally expressive in his speeches compared to the Democratic representative.

```python
# plot distribution of polarity and subjectivity
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes = axes.ravel()
for harris, trump, sentiment, i in zip((harris_pol, harris_sub), 
                                     (trump_pol, trump_sub), 
                                     ('Polarity', 'Subjectivity'),
                                   range(2)):
    # plot distribution of scores
    # sns.kdeplot(harris, label='Harris', color='b', alpha=0.5, fill=True, ax=axes[i])
    # sns.kdeplot(trump, label='Trump', color='r', alpha=0.5, fill=True, ax=axes[i])
    sns.ecdfplot(harris, label='Harris', color='b', ax=axes[i])
    sns.ecdfplot(trump, label='Trump', color='r', ax=axes[i])
    # add labels and titles
    axes[i].set_xlabel(f'{sentiment} score')
    axes[i].set_ylabel('CDF')
    # add annotation box for results of KS-test
    ks = stats.ks_2samp(harris, trump)
    d_stat, p_val = ks.statistic, ks.pvalue
    # Add a text box with the D-statistic and p-value
    textstr = f'D-statistic: {d_stat:.4f}\np-value: {p_val:.4f}'
    props = dict(boxstyle='round', facecolor='none', alpha=0.5, edgecolor='none')
    axes[i].text(0.03, 0.8, textstr, transform=axes[i].transAxes, fontsize=10,
        verticalalignment='top', bbox=props)
    axes[i].legend()
    
# fig.suptitle('Polarity and Subjectivity cumulative distribution', fontsize=16)
plt.tight_layout()
plt.savefig('subjectivity_polarity_cumulative_distribution.png')
plt.show()
```

Polarity scores distribution of Harris is much more leptokurtic than Trump's, with high peak at around 0, which confirms our ealier claim that her sentiment are more neutral and does not show as much variation as Trump.

Interestingly, the subjectivity score of both candidates follow a bimodal distribution. While the first (lower) peak is at around 0, Harris' second (higher) peak is at around 0.3 to 0.4 while Trump's second peak is at a higher score (about 0.6). This shows that, while both candidates are subjective in the debate, Trump is overall more subjective than Harris.

### 4.2. Sentiment intensity over time

```python
# function to extract sentiment intensity speech by speech
def get_sentiment_intensity_over_time(speech_list):
    positive = []
    negative = []
    neutral = []
    compound = []
    
    for s in speech_list:
        polarity_scores = sia.polarity_scores(s)
        positive.append(polarity_scores['pos'])
        negative.append(polarity_scores['neg'])
        neutral.append(polarity_scores['neu'])
        compound.append(polarity_scores['compound'])
    return positive, negative, neutral, compound
```

```python
harris_pos, harris_neg, harris_neu, harris_compound = get_sentiment_intensity_over_time(harris_speech_cleaned)
trump_pos, trump_neg, trump_neu, trump_compound = get_sentiment_intensity_over_time(trump_speech_cleaned)
```

```python
# plot sentiment intensity over time
for harris, trump, sentiment in zip((harris_pos, harris_neg, harris_neu, harris_compound), 
                                     (trump_pos, trump_neg, trump_neu, trump_compound), 
                                     ('positive', 'negative', 'neutral', 'compound')):
    fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 5))

    ax[0].plot(harris, color='b', label='Harris')
    ax[0].set_ylabel('Sentiment intensity')
    ax[0].set_xlabel('Speech')
    ax[0].set_title(f'Sentiment strength over time ({sentiment})')
    ax[0].legend(loc='upper right')
    
    ax[1].plot(trump, color='red', label='Trump')
    # ax[1].set_ylim(0, 1)
    ax[1].set_ylabel('Sentiment intensity')
    ax[1].set_xlabel('Speech')
    ax[1].legend(loc='upper right')

    if sentiment == 'compound':
        ax[0].set_ylim(-1.1, 1.1)
        ax[1].set_ylim(-1.1, 1.1)
    else:
        ax[0].set_ylim(-0.1, 1.1)
        ax[0].set_ylim(-0.1, 1.1)
        
    plt.tight_layout()
    plt.savefig(f'sentiment_strength_over_time_{sentiment}.png')
    plt.show()
```

```python
# plot distribution of sentiment intensity scores
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
axes = axes.ravel()
for harris, trump, sentiment, i in zip((harris_pos, harris_neg, harris_neu, harris_compound), 
                                     (trump_pos, trump_neg, trump_neu, trump_compound), 
                                     ('positive', 'negative', 'neutral', 'compound'),
                                   range(4)):
    # plot distribution of sentiment intensity scores
    # sns.kdeplot(harris, label='Harris', color='b', alpha=0.5, fill=True, ax=axes[i])
    # sns.kdeplot(trump, label='Trump', color='r', alpha=0.5, fill=True, ax=axes[i])
    sns.ecdfplot(harris, label='Harris', color='b', ax=axes[i])
    sns.ecdfplot(trump, label='Trump', color='r', ax=axes[i])
    # add labels and titles
    axes[i].set_xlabel(f'{sentiment} score')
    axes[i].set_ylabel('CDF')
    # add annotation box for results of KS-test
    ks = stats.ks_2samp(harris, trump)
    d_stat, p_val = ks.statistic, ks.pvalue
    # Add a text box with the D-statistic and p-value
    textstr = f'D-statistic: {d_stat:.4f}\np-value: {p_val:.4f}'
    props = dict(boxstyle='round', facecolor='none', alpha=0.5, edgecolor='none')
    axes[i].text(0.03, 0.8, textstr, transform=axes[i].transAxes, fontsize=10,
        verticalalignment='top', bbox=props)
    axes[i].legend()
    
    
# fig.suptitle('Sentiment intensity distribution', fontsize=16)
plt.tight_layout()
plt.savefig('sentiment_intensity_cumulative_distribution.png')
plt.show()
```

### 5.8. Summarize and plotting all the results by topic

```python
# polarity and subjectivity
measures = ['polarity', 'subjectivity']
dfs_ = []
for i in measures:
    df = pd.DataFrame({topic: {candidate: polarity_subjectivity_by_topic[topic][candidate][i] 
                                    for candidate in polarity_subjectivity_by_topic[topic]} for topic in polarity_subjectivity_by_topic})
    df = df.T  # Transpose to have first-level keys as index and candidates as columns
    df.columns = [f'harris_{i}', f'trump_{i}']
    # append to the list of df
    dfs_.append(df)

# merge the dfs into one table for polarity and subjectivity
pol_sub_df = pd.concat(dfs_, axis=1)
pol_sub_df
```

```python
# sentiment intensity
measures = ['neg', 'neu', 'pos', 'compound']
dfs_ = []
for i in measures:
    df = pd.DataFrame({topic: {candidate: sentiment_intensity_by_topic[topic][candidate][i] 
                                    for candidate in sentiment_intensity_by_topic[topic]} for topic in sentiment_intensity_by_topic})
    df = df.T  # Transpose to have first-level keys as index and candidates as columns
    df.columns = [f'harris_{i}', f'trump_{i}']
    # append to the list of df
    dfs_.append(df)

# merge the dfs into one table for sentiment intensity
sentiment_intensity_df = pd.concat(dfs_, axis=1)
sentiment_intensity_df
```

```python
# sentiment_df = pd.concat(pol_sub_df, sentim)
sentiment_df = pd.concat([pol_sub_df, sentiment_intensity_df], axis=1)
sentiment_df = sentiment_df.round(4)

sentiment_df.to_csv('sentiment_by_topic.csv')
```

```python
# emotion score
measures = ['anticipation', 'positive', 'joy', 'trust', 'anger', 'fear', 'negative', 'sadness', 'surprise', 'disgust']
dfs_ = []
for i in measures:
    df = pd.DataFrame({topic: {candidate: emotion_score_by_topic[topic][candidate][i] 
                                    for candidate in emotion_score_by_topic[topic]} for topic in emotion_score_by_topic})
    df = df.T  # Transpose to have first-level keys as index and candidates as columns
    df.columns = [f'harris_{i}', f'trump_{i}']
    # append to the list of df
    dfs_.append(df)

# merge the dfs into one table for emotion score
emotion_df = pd.concat(dfs_, axis=1)
emotion_df
```

```python
t = emotion_df.loc['economy'][[col for col in emotion_df.columns if 'harris_' in col]]
t.index = [i.replace('harris_', '') for i in t.index]
t
```

```python
# let's plot the emotion scores on all topic 
fig, ax = plt.subplots(nrows=2, ncols=4, figsize=(20, 8))
ax = ax.ravel()
topics = emotion_df.index
# print(topics)
for i, t in enumerate(topics):
    # get data for plotting
    harris = emotion_df.loc[t][[col for col in emotion_df.columns if 'harris_' in col]]
    harris.index = [i.replace('harris_', '') for i in harris.index]
    
    trump = emotion_df.loc[t][[col for col in emotion_df.columns if 'trump_' in col]]
    trump.index = [i.replace('trump_', '') for i in trump.index]

    scores = pd.concat([harris, trump], axis=1)
    scores.columns = ['Harris', 'Trump']

    scores.sort_index(ascending=False).plot(kind='barh', color=['blue', 'red'], ax=ax[i])
    ax[i].set_title(t)
    ax[i].set_xlim(0, 0.4)

# plt.suptitle('Emotion scores (normalized) by topic', fontsize=18, y=1)
plt.tight_layout()
plt.savefig('emotion_scores_by_topic.png')
plt.show()
```
